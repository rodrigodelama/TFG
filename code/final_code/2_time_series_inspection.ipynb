```python
import pandas as pd
import matplotlib.pyplot as plt

# Debug parameter
DEBUG = False

if DEBUG:
    print("Debug mode is ON. Detailed output will be printed.")

# Load Database
csv_hour_file = '../data/ta_metrics/final_price_ta_metrics.csv'
df = pd.read_csv(csv_hour_file, parse_dates=['Datetime'])

# Keep only Datetime and MarginalES
df = df[['Datetime', 'MarginalES']]
```

```python
# Visualize the time series with matplotlib

plt.figure(figsize=(14, 7))
# Use UC3M sRGB color RGB(0, 14, 120)
plt.plot(df['Datetime'], df['MarginalES'], color='darkblue')
plt.title('MarginalES Energy Price Over Time at 14:00H')
plt.xlabel('Datetime')
plt.ylabel('Energy Price €/MWh')
plt.grid(alpha=0.3)

# Save the plot as a PNG file BEFORE showing it
plt.savefig('marginal_energy_price_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Data frame inspection+

# Max price
max_price = df['MarginalES'].max()
# Mean price
mean_price = df['MarginalES'].mean()
# Median price
median_price = df['MarginalES'].median()
# Mode price
mode_price = df['MarginalES'].mode()[0] if not df['MarginalES'].mode().empty else None


# Print the results
print(f"Max Price: {max_price:.2f} €/MWh")
print(f"Mean Price: {mean_price:.2f} €/MWh")
print(f"Median Price: {median_price:.2f} €/MWh")
print(f"Mode Price: {mode_price:.2f} €/MWh" if mode_price is not None else "Mode Price: No mode found")
```

```python
# Draw value distribution including mark for average value and mode value
plt.figure(figsize=(14, 7))
# Use UC3M sRGB color RGB(0, 14, 120) for the histogram
plt.hist(df['MarginalES'], bins=50, color="darkblue", edgecolor='white')

plt.axvline(df['MarginalES'].mean(), color='crimson', linestyle='dashed', linewidth=2, label='Mean Price - {:.2f} €/MWh'.format(mean_price))
plt.axvline(df['MarginalES'].median(), color='gold', linestyle='dashed', linewidth=2, label='Median Price - {:.2f} €/MWh'.format(median_price))
plt.axvline(df['MarginalES'].mode()[0], color='forestgreen', linestyle='dashed', linewidth=2, label='Mode Price - {:.2f} €/MWh'.format(mode_price) if mode_price is not None else 'Mode Price - No mode found')

plt.title('Distribution of MarginalES Energy Prices at 14:00H')
plt.xlabel('Energy Price €/MWh')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.legend()
# Save the histogram as a PNG file BEFORE showing it
plt.savefig('marginal_energy_price_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```
